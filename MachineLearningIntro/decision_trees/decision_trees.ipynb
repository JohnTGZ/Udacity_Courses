```python
import numpy as np
# Entropy 

bucket = [8,3,2]
total_balls = sum(bucket)
entropy = 0

for b in bucket:
    entropy -= (b/total_balls) * np.log2((b/total_balls))


print("total entropy: ", entropy)
```

```python
import numpy as np
def calcEntropy(list):
    total_elem = sum(list)

    entropy = 0
    for val in list:
        entropy -= (val/total_elem) * np.log2(val/total_elem)
        # print(entropy)

    return entropy
```

```python
from sklearn.linear_model import LinearRegression
import pandas as pd

def getShape(array, name):
    print("shape of ", name, ": ", array.shape)

df = pd.read_csv('ml-bugs.csv')

#Print head of dataframe
# print(df.head)

#Extract rows based on column value
#1. https://stackoverflow.com/questions/42128784/extract-rows-from-pandas-dataframe-based-on-column-value
#2. https://stackoverflow.com/questions/17424182/extracting-all-rows-from-pandas-dataframe-that-have-certain-value-in-a-specific

#Intial entropy
group_a_list = [len(df[df['Species'] == 'Mobug']), len(df[df['Species'] != 'Mobug'])] 

print (f"Mobug: {group_a_list[0]}, No Mobug: {group_a_list[1]}")
initial_entropy = calcEntropy(group_a_list)

print (f"Initial entropy: {initial_entropy}")

###################################
#Extract only bugs with specified color
###################################
decision_param = ['Brown', 'Blue', 'Green']

for decision in decision_param:
        print(f"===============")
        print(f"{decision}")
        grp_a = df[df['Color'] == decision]
        grp_b = df[df['Color'] != decision]
        print(f"        No. of {decision} bugs: {len(grp_a)}, \
                and non-{decision} bugs:{len(grp_b)}", )

        grp_a_list = [len(grp_a[grp_a['Species'] == 'Mobug']), len(grp_a[grp_a['Species'] != 'Mobug'])] 
        grp_b_list = [len(grp_b[grp_b['Species'] == 'Mobug']), len(grp_b[grp_b['Species'] != 'Mobug'])]

        print(f"        {decision}: Mobug: {grp_a_list[0]}, \
                and Lobug:{grp_a_list[1]}", )
        print(f"        NON-{decision}: Mobug: {grp_b_list[0]}, \
                and Lobug:{grp_b_list[1]}", )


        group_a_entropy = calcEntropy(grp_a_list)
        group_b_entropy = calcEntropy(grp_b_list)

        total_grp_a = sum(grp_a_list)
        total_grp_b = sum(grp_b_list)
        average_entropy = (group_a_entropy*total_grp_a + group_b_entropy*total_grp_b)/24
        
        info_gain = initial_entropy - average_entropy
        print (f"       For {decision}: Info_gain: {initial_entropy} - {average_entropy} = {info_gain}")

###################################
#Extract only bugs with length less than specified
###################################

decision_param = [17.0, 20.0]

for decision in decision_param:
        print(f"{decision}")

        grp_a = df[df['Length (mm)'] < decision]
        grp_b = df[df['Length (mm)'] >= decision]
        print(f"        No. of <{decision} bugs: {len(grp_a)}, \
                and >={decision} bugs:{len(grp_b)}", )

        grp_a_list = [len(grp_a[grp_a['Species'] == 'Mobug']), len(grp_a[grp_a['Species'] != 'Mobug'])] 
        grp_b_list = [len(grp_b[grp_b['Species'] == 'Mobug']), len(grp_b[grp_b['Species'] != 'Mobug'])]

        print(f"        {decision}: Mobug: {grp_a_list[0]}, \
                and Lobug:{grp_a_list[1]}", )
        print(f"        NON-{decision}: Mobug: {grp_b_list[0]}, \
                and Lobug:{grp_b_list[1]}", )

        group_a_entropy = calcEntropy(grp_a_list)
        group_b_entropy = calcEntropy(grp_b_list)

        total_grp_a = sum(grp_a_list)
        total_grp_b = sum(grp_b_list)
        average_entropy = (group_a_entropy*total_grp_a + group_b_entropy*total_grp_b)/24
        
        info_gain = initial_entropy - average_entropy
        print (f"       For {decision}: Info_gain: {initial_entropy} - {average_entropy} = {info_gain}")
```
